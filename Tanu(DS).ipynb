```python
# prompt: can you find out basic stats information on my dataframe df , calculate measurements such as mean , median, mode, variance, std, range , skewness , kurtosis

import pandas as pd

# Assuming your dataframe is named 'df' and has a column named 'salary'

# Calculate mean
mean_salary = df['salary'].mean()

# Calculate median
median_salary = df['salary'].median()

# Calculate mode
mode_salary = df['salary'].mode()[0]  # Access the first mode if multiple exist

# Calculate variance
variance_salary = df['salary'].var()

# Calculate standard deviation
std_salary = df['salary'].std()

# Calculate range
range_salary = df['salary'].max() - df['salary'].min()

# Calculate skewness
skewness_salary = df['salary'].skew()

# Calculate kurtosis
kurtosis_salary = df['salary'].kurt()


print(f"Mean: {mean_salary}")
print(f"Median: {median_salary}")
print(f"Mode: {mode_salary}")
print(f"Variance: {variance_salary}")
print(f"Standard Deviation: {std_salary}")
print(f"Range: {range_salary}")
print(f"Skewness: {skewness_salary}")
print(f"Kurtosis: {kurtosis_salary}")
```

```python
# prompt: can you create a distplot for all the columns in data frames in df1

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df1 is already loaded as in your provided code
for col in df1.columns:
    plt.figure()  # Create a new figure for each column
    sns.distplot(df1[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```
